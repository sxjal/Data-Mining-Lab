```python
import pandas as pd
df2=pd.read_csv('hotel_bookings.csv')

print(df2.shape) #row,col
sumres = df2['agent'].sum(skipna = True) #gets the sum of the agent column in the dataset
countres = df2['agent'].count() #get the count of all the available cells for average



avg = sumres/countres #avg of the data set
#will use this data to fill the cells where the data is null in the agent column

print("sum",sumres)
print("count",countres)
print(avg)
```

```python
df2.head()
#get the columns which are null
```

```python
df2.loc[df2['agent'].isnull(), 'agent'] = avg
```
